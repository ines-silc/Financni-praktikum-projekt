```python
import time #Za primerjavo časovne zahtevnosti
import networkx as nx #Za definiranje grafov
import matplotlib.pyplot as mpl #Za risanje grafov
import itertools #Za urejanje seznamov z nabori
import random
```

```python
prvotna_pot = [0, 1, 2, 3, 4]
izhodisce = prvotna_pot[0]
prvotna_pot.append(izhodisce)

matrika = tg.testna_matrika2
zacetni_cas = time.clock()
resitev2opt1 = pr.dva_opt(matrika, prvotna_pot)
koncni_cas = time.clock()
cas = koncni_cas - zacetni_cas
print('Čas delovanja:', cas, '  pot:', prvotna_pot, '  dolzina poti:', pr.cena_poti(matrika, resitev2opt1[0]))
print('Najboljša rešitev je', resitev2opt1)

primer1 = pr.pretvori_matriko_v_graf(matrika)
pos = nx.circular_layout(primer1)
nx.draw(primer1, pos = pos, with_labels = True)
nx.draw_networkx_edge_labels(primer1, pos = pos, labels = nx.get_edge_attributes(primer1, 'weight').values())
mpl.savefig("primeri/primer1.png", format="PNG", dpi = 1000)
mpl.show()

najboljsa_pot = resitev2opt1[0]
print(najboljsa_pot)

primer1_2opt = pr.pretvori_matriko_v_graf(pr.pretvori_resitev_v_matriko(najboljsa_pot, matrika))
nx.draw(primer1_2opt, pos = pos, with_labels = True)
nx.draw_networkx_edge_labels(primer1_2opt, pos = pos, labels = nx.get_edge_attributes(primer1_2opt, 'weight'))
mpl.savefig("primeri/primer1_2opt.png", format="PNG", dpi = 1000)
mpl.show()

prvotna_pot = [0, 1, 2, 3, 4]

zacetni_cas = time.clock()
pot = pr2.tri_opt(matrika, prvotna_pot)
koncni_cas = time.clock()
cas = koncni_cas - zacetni_cas
cikel = pot[:]
cikel.append(pot[0])
print('Čas delovanja:', cas, '  pot:', cikel, '  dolzina poti:', pr.cena_poti(matrika, cikel))

primer1_3opt = pr.pretvori_matriko_v_graf(pr.pretvori_resitev_v_matriko(cikel, matrika))
nx.draw(primer1_3opt, pos = pos, with_labels = True)
nx.draw_networkx_edge_labels(primer1_3opt, pos = pos, labels = nx.get_edge_attributes(primer1_3opt, 'weight'))
mpl.savefig("primeri/primer1_3opt.png", format="PNG", dpi = 1000)
mpl.show()

zacetni_cas = time.clock()
primer1_lk = lk.Lin_Kernighan(matrika, prvotna_pot)
koncni_cas = time.clock()
cas = koncni_cas - zacetni_cas
cikel = primer1_lk
cikel.append(cikel[0])
print('Čas delovanja:', cas, '  pot:', cikel, '  dolzina poti:', pr.cena_poti(matrika, cikel))

primer1_lk = lk.pretvori_matriko_v_graf(lk.pretvori_resitev_v_matriko(cikel, matrika))
nx.draw(primer1_lk, pos = pos, with_labels = True)
nx.draw_networkx_edge_labels(primer1_lk, pos = pos, labels = nx.get_edge_attributes(primer1_lk, 'weight'))
mpl.savefig("primeri/primer1_lk.png", format="PNG", dpi = 1000)
mpl.show()
```
