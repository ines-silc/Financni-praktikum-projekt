## 2-opt

```python
import program as pr
import testni_graf as tg
```

```python
poti_in_cene = []
for i in range(0, len(list(itertools.permutations(range(5)))), 30):
    prvotna_pot = list(list(itertools.permutations(range(5)))[i])
    izhodisce = prvotna_pot[0]
    prvotna_pot.append(izhodisce)   
    (a, b) = pr.dva_opt(tg.testna_matrika2, prvotna_pot)
    poti_in_cene.append((a, b))
```

```python
prvotna_pot = [0, 1, 3, 4, 2]
izhodisce = prvotna_pot[0]
prvotna_pot.append(izhodisce)   
print(pr.dva_opt(tg.testna_matrika2, prvotna_pot))
```

```python
poti_in_cene.sort(key=lambda elem: elem[1])
print('Vse poti so', poti_in_cene)
print('Najboljša rešitev je', poti_in_cene[0])
```

```python
tg2 = pr.pretvori_matriko_v_graf(tg.testna_matrika2)
nx.draw(tg2, pos = nx.circular_layout(tg2), with_labels = True)
nx.draw_networkx_edge_labels(tg2, pos = nx.circular_layout(tg2), labels = nx.get_edge_attributes(tg2, 'weight').values())
mpl.show()

najboljsa_pot = poti_in_cene[0][0]
```

```python
rtg3 = pr.pretvori_matriko_v_graf(pr.pretvori_resitev_v_matriko(najboljsa_pot, tg.testna_matrika2))
nx.draw(rtg3, pos = nx.circular_layout(rtg3), with_labels = True)
nx.draw_networkx_edge_labels(rtg3, pos = nx.circular_layout(rtg3), labels = nx.get_edge_attributes(rtg3, 'weight'))
mpl.show()
```

```python
primer2 = tg.matrika1
n1 = 10
print(primer2)
```

```python
poti_in_cene = []
for i in range(0, len(list(itertools.permutations(range(n1)))), 10000):
    prvotna_pot = list(list(itertools.permutations(range(n1)))[i])
    izhodisce = prvotna_pot[0]
    prvotna_pot.append(izhodisce)   
    (a, b) = pr.dva_opt(primer2, prvotna_pot)
    poti_in_cene.append((a, b))
    
poti_in_cene.sort(key=lambda elem: elem[1])
print('Vse poti so', poti_in_cene)
print('Najboljša rešitev je', poti_in_cene[0])
```

## 3-opt

```python
import tri_opt as pr2
```

```python
primer3 = tg.testna_matrika2

resitve = []
for i in range(0, len(list(itertools.permutations(range(5))))):
    prvotna_pot = list(list(itertools.permutations(range(5)))[i])
    pot = pr2.tri_opt(primer3, prvotna_pot)
    #print('Rešitev je', pot)
    cikel = pot
    cikel.append(pot[0])
    cena = pr.cena_poti(primer3, cikel)
    #print('Cikel je', cikel, ', in njegova cena je', cena)
    resitve.append((pot, cena))

resitve.sort(key=lambda elem: elem[1])
#print('Vse poti so', resitve)
print('Najboljša rešitev je', resitve[0])
```

```python
pr3 = pr.pretvori_matriko_v_graf(primer3)
nx.draw(pr3, pos = nx.circular_layout(pr3), with_labels = True)
nx.draw_networkx_edge_labels(pr3, pos = nx.circular_layout(pr3), labels = nx.get_edge_attributes(pr3, 'weight').values())
mpl.savefig("primer_3_opt.png", format="PNG")
mpl.show()

rpr3 = pr.pretvori_matriko_v_graf(pr.pretvori_resitev_v_matriko(resitve[0][0], primer3))
nx.draw(rpr3, pos = nx.circular_layout(rpr3), with_labels = True)
nx.draw_networkx_edge_labels(rpr3, pos = nx.circular_layout(rpr3), labels = nx.get_edge_attributes(rpr3, 'weight'))
mpl.savefig("resitev_primer_3_opt.png", format="PNG")
mpl.show()
```
